```python
%matplotlib inline 
```

```python
yasListesi=[13,16,20,43,46,67,85,90,99,100]
```

```python
kiloListesi=[40,50,60,70,80,90,100,70,50,40]
```

```python
numpyYasListesi=np.array(yasListesi)
numpyKiloListesi=np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Yaş Kilo Değişimi")
plt.show()
```

```python
numpyDizisi1=np.linspace(-20,20,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2=numpyDizisi1**2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"r--")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
```

```python
plt.subplot(1,2,1) # 1 sıra 2 kolon 1. yi çiziyorum
plt.plot(numpyDizisi1,numpyDizisi2,"r--")
plt.subplot(1,2,2) # 1 sıra 2 kolon 2. yi çiziyorum
plt.plot(numpyDizisi2,numpyDizisi1,"b--")
# 2 kolon demek yan yana 2 grafik olacak demek
```

```python
benimFigure=plt.figure() # boş bir figür yani grafiğin çizileceği şey oluşturdum
figureAxes=benimFigure.add_axes([0.2,0.2,0.9,0.9]) # tablonun başlama bitme kordinatları gibi bir şey ileride netleşecek
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g--")
figureAxes.set_xlabel("X Ekseni")
figureAxes.set_ylabel("Y Ekseni")
figureAxes.set_title("Grafik")
```

```python
figure2=plt.figure()
eksen1=figure2.add_axes([0.1,0.1,0.7,0.9]) # sağdaki ikili grafiğin çizildiği şeyin boyutları 
eksen2=figure2.add_axes([0.3,0.6,0.3,0.3]) # soldaki ikili kordinat

eksen1.plot(numpyDizisi1,numpyDizisi2,"r")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik")

eksen2.plot(numpyDizisi1,numpyDizisi2,"r")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Ana Grafik")
```

```python
(benimFigure , BenimEksen)=plt.subplots()
#yukarıdaki gibi eksen ve figure ayrı ayrı oluşturmaktansa burada tek satırda oluşturmuş olduk .
BenimEksen.plot(numpyDizisi1,numpyDizisi2,"g")
```

```python
(benimFigure , BenimEksenler)=plt.subplots(nrows=1,ncols=2)
for eksen in BenimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,color="#5c033d",alpha=0.3)
    # alpha saydamlık
plt.tight_layout() # grafiklerin birbirlerine olan mesafesi için
```

```python
yeniFigure=plt.figure() 
# dpi çözünürlük verir , burada da verebilirsin kayıt esnasında da verebilirsin
# parantez içine figuresize=(6,6) gibi verebilirsin şu anda otomatik ayarlıyor 

yeniEksen=yeniFigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi2,label="numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label="numpyDizisi ** 3")
yeniEksen.legend(loc=2)
```

```python
yeniFigure.savefig("benimFigure.png",dpi=161)
```

```python
(yeniigur,yeniEksen)=plt.subplots()
yeniEksen.plot(numpyDizisi1,numpyDizisi1+2,color="blue",linewidth=5.0)
yeniEksen.plot(numpyDizisi1,numpyDizisi1+5,color="red")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+8,color="#5c033d",linestyle="--")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+9,color="#5c033d",linestyle="-.")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+10,color="#5c033d",linestyle=":")


yeniEksen.plot(numpyDizisi1,numpyDizisi1+13,color="#000000",marker="*",markerfacecolor="red",markersize=10)
```
