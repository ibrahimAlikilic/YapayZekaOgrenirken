Şimdi görüntünün yansılamarını elde edeceğiz
#np.flip(img, ? ) img den sonra 3 tane değer girebiliyoruz0,1,2
* 0 : x eksenine göre yansıma
* 1 : y eksenine göre
* 2 : aşşağıda daha anlaşılır

```python
plt.subplot(1,3,1)
plt.title("np.flip(img,0)")
plt.imshow(np.flip(img,0))

plt.subplot(1,3,2)
plt.title("np.flip(img,1)")
plt.imshow(np.flip(img,1))


plt.subplot(1,3,3)
plt.title("np.flip(img,2)")
plt.imshow(np.flip(img,2))
```
