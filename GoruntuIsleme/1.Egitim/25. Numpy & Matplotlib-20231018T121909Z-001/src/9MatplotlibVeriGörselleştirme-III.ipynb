```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(3)
```

```python
plt.plot(x,[y**2 for y in x])
plt.plot(x,[y**3 for y in x])
plt.plot(x,[y*2 for y in x])
plt.plot(x,[y*5.2 for y in x])
# eklemeler yapıyoruz
# analizde kolaylık olsun diye ızgaralar ekliyelim
plt.grid(True)
plt.show()
```

```python
plt.plot(x,[y**2 for y in x])
plt.plot(x,[y**3 for y in x])
plt.plot(x,[y*2 for y in x])
plt.plot(x,[y*5.2 for y in x])
print(plt.axis()) # yukarıdaki grafikte görüldüğü gibi grafiğin başlangıç ve bitiş değerleri tabloda yazmıyor bunları öğrenmek için bu fonksiyonu kullanıyoruz
plt.show()
```

```python
plt.plot(x,[y**2 for y in x])
plt.plot(x,[y**3 for y in x])
plt.plot(x,[y*2 for y in x])
plt.plot(x,[y*5.2 for y in x])
print(plt.axis([0,2,0,10])) # şimdi de öğrenmiş olduğum bu noktaları grafiĞİN başlangıç ve bitiş olarak yazıcaz İLK 2 DEĞER X İN DİĞER 2 DEĞER Y NİN
plt.show()
```

```python
# şimdi de yapmış olduğumuz çizimde hangi çizgi hangi fonksiyona ait
plt.plot(x,[y**2 for y in x])
plt.plot(x,[y**3 for y in x])
plt.plot(x,[y*2 for y in x])
plt.plot(x,[y*5.2 for y in x])
plt.legend(['x**2','x**3','x*2','x*5.2'])
plt.show()
```

```python
# burada da yerini ayarlayacağız
plt.plot(x,[y**2 for y in x])
plt.plot(x,[y**3 for y in x])
plt.plot(x,[y*2 for y in x])
plt.plot(x,[y*5.2 for y in x])
plt.legend(['x**2','x**3','x*2','x*5.2'],loc="upper center")
plt.show()
```

```python
# burada da yerini ayarlayacağız
plt.plot(x,[y**2 for y in x])
plt.plot(x,[y**3 for y in x])
plt.plot(x,[y*2 for y in x])
plt.plot(x,[y*5.2 for y in x])
plt.legend(['x**2','x**3','x*2','x*5.2'],loc="lower left") # mantığında.
plt.show()
```

```python
# eksenlere ad koyuyoruz
plt.plot(x,[y**2 for y in x])
plt.plot(x,[y**3 for y in x])
plt.plot(x,[y*2 for y in x])
plt.plot(x,[y*5.2 for y in x])
plt.xlabel("x=np.arange(3)")
plt.ylabel("y=f(x)")
plt.show()
```
