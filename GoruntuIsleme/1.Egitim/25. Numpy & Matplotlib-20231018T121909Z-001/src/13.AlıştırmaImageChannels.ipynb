RGB değerlerini değişken üzerinde tutucaz

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path="C:\\Users\\ibo_m\\Desktop\\Yapay Zeka\\Görüntü işleme\\25. Numpy & Matplotlib-20231018T121909Z-001\\input\\13.2 map.jpeg.jpeg"
img=plt.imread(path)
```

```python
r=img[:,:,0] # tüm x ve y lerdeki red değerlerini tut
g=img[:,:,1]
b=img[:,:,2]
```

```python
output=[img,r,g,b,]
```

```python
# şimdi birden fazla resmi for döngüsüyle tek resimde göstericez
titles = ["Image","Red","Green","Blue"]
for i in range(4):
    plt.subplot(2,2,i+1) # alt grafikler oluşturma komutu
                #kaç satır , sütun , grafiklerin nereye yerleşeceği
    plt.axis("off") # oluşan her geafik için eksenleri yok ettik
    plt.title(titles[i])
    if i ==0:
        plt.imshow(output[i])
    else:
        plt.imshow(output[i],cmap='gray') # aydınlık olan yerlerde yoğun demek gri ye çevirmeyince düzgün vermiyor
    plt.show()            
```

```python
# renkleri üst üste yığarak orijinal formata çevirebiliyoruz
output = np.dstack((r,g,b))
plt.imshow(output)
plt.show()
```
