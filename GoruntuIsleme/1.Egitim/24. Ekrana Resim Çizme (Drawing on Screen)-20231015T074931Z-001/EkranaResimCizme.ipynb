```python
while 1:
    ret,frame=cap.read()
    frame=cv2.flip(frame,1)
    #ilk önce mavi nesnemizi yakalayalım
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    # şimdi yukarıda oluşturduğum şeyleri burada da yapıyorum çünkü frameler üzerinde de oluşturmak istiyoruz
    frame=cv2.rectangle(frame,(40,1),(140,65),(0,0,0),2) # bunun içi boş diğerlerininki dolu
    frame=cv2.rectangle(frame,(160,1),(255,65),colors[0],-1)
    frame=cv2.rectangle(frame,(275,1),(370,65),colors[1],-1)
    frame=cv2.rectangle(frame,(390,1),(485,65),colors[2],-1)
    frame=cv2.rectangle(frame,(505,1),(600,65),colors[3],-1)
    # şimdi yazıları yazacağız
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame,"CLEAR ALL",(49,33),font,0.5,(0,0,0),2,cv2.LINE_AA)
    cv2.putText(frame,"BLUE",(185,33),font,0.5,(255,255,255),2,cv2.LINE_AA)
    cv2.putText(frame,"GREEN",(298,33),font,0.5,(255,255,255),2,cv2.LINE_AA)
    cv2.putText(frame,"RED",(420,33),font,0.5,(255,255,255),2,cv2.LINE_AA)
    cv2.putText(frame,"YELLOW",(520,33),font,0.5,(255,255,255),2,cv2.LINE_AA)
    
    if ret is False: # video bittiğinde ya da yanlış okumada döngüden çık demek
        break
    # şimdi mask uyguluyoruz
    mask=cv2.inRange(hsv,lower_blue,upper_blue) # bak hata yaptın ferame ye değil hsv formatında olduğundan hsv ye uygulayacaksın , ezbere gitme !
    # şimdi framedeki karıncalanmaları yok edeceğiz
    mask=cv2.erode(mask,(5,5),iterations=2)
    mask=cv2.morphologyEx(mask,cv2.MORPH_OPEN,(5,5)) # açıklamaları google dosyanda var
    mask=cv2.dilate(mask,(5,5),iterations=1)
    # contours arama aşamasına geldik
    contours,_=cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    # tutumuş olduğum nesnenin merkezine çember oluşturucam ne zmn çember düğmenin içine girer o zmn düğmeyi kullan diyeceğiz
    center=None
    # artık renkleri alacağız
    if len(contours)>0:
        max_contours = sorted(contours, key = cv2.contourArea, reverse=True)[0]# reverse=True büyükten küçüğe demek
        # şimdi contouru sınırlayan bir çember çizeceğiz
        ((x,y),radius)=cv2.minEnclosingCircle(max_contours)
        # şimdi bu çemberi çizelim
        cv2.circle(frame,(int(x),int(y)),int(radius),(255,0,255),3)
        # şimdi çemberin merkezini buluyoruz
        M=cv2.moments(max_contours)
        try: # baktım mavi olmayınca hata veriyor hata sebebi de burası normal olarak merkez bulamayınca hata veriyor o yüzden try except bloğuna aldım ve oldu
            center=(int(M["m10"]/M["m00"]),int(M["m01"]/M["m00"])) # bunun ayrıntılarını önceki derslerde yapmıştım ; sonuçlar int olmalı yoksa hata veriyor
        except:
            continue
        # şimdi center merkezime göre renk değiştireceğim
        if center[1]<65:
            if 40<=center[0]<=140:
                # önceden tanımladığım syf değişkenlerini kullanıyorum
                blue_points = [deque(maxlen=512)]
                green_points = [deque(maxlen=512)]
                red_points = [deque(maxlen=512)]
                yellow_points = [deque(maxlen=512)]

                blue_index=0
                green_index=0
                red_index=0
                yellow_index=0
                # şimdi ekranı temizliyorum bembeyaz yapıyorum
                # butonların altından itibaren temizleticem çünkü uralara işlem yapmayacağım
                paintWindow[67:,:,:]=255 # y , x , channel değeri 
            elif 160<=center[0]<=255:
                color_index = 0

            elif 275<=center[0]<=370:
                color_index = 1

            elif 390<=center[0]<=485:
                color_index = 2

            elif 505<=center[0]<=600:
                color_index = 3
        # 65ten büyük yani çizim kısmındaysa
        else:
            if color_index == 0:
                blue_points[blue_index].appendleft(center)# center gezdiği yerleri soldan buraya ekle yani noktaları birleştiriyoruz
            elif color_index == 1:
                green_points[green_index].appendleft(center)
                
            elif color_index == 2:
                red_points[red_index].appendleft(center)
                
            elif color_index == 3:
                yellow_points[yellow_index].appendleft(center)
    else:
        # burada pointleri eklicez
        blue_points.append(deque(maxlen=512))
        blue_index+=1
        green_points.append(deque(maxlen=512))
        green_index+=1
        red_points.append(deque(maxlen=512))
        red_index+=1
        yellow_points.append(deque(maxlen=512))
        yellow_index+=1
        
        # şimdi bunları poins ime atayacağım
    points=[blue_points,green_points,red_points,yellow_points]
    # for döngüsüyle bu değerleri alacağım
    for i in range(len(points)):
        for j in range(len(points[i])):
            for k in range(1,len(points[i][j])):
                # çizim yapmadan önce hata alamamk için listelerin içinin boş olmadığına yani noktalar olduğuna emin olmam lazım
                # bu yüzden karar yapısı oluşturacağız
                if points[i][j][k-1] is None or points[i][j][k] is None :  # points[i][j][k-1] başlangıç noktasıdır , points[i][j][k] bitiş noktasıdır
                    continue
                cv2.line(frame,points[i][j][k-1], points[i][j][k], colors[i], 2)
                cv2.line(paintWindow,points[i][j][k-1], points[i][j][k], colors[i], 2)
        
    cv2.imshow("frame",frame)
    cv2.imshow("Paint",paintWindow) # paintWindow için Paint penceresi oluşturmuştum ve orada açılmasını istiyorum
    cv2.imshow("mask",mask)
    if cv2.waitKey(3)&0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
