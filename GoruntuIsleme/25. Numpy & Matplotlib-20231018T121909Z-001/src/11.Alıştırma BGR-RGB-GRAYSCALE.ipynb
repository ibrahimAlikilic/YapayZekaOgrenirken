```python
import cv2
import matplotlib.pyplot as plt
```

```python
# isimlendirmelerden dolayı bu örnek hata verdi en baştan itibaren isimleri değiştirmek istemedim
img=cv2.imread("11.2 smile.jpg.jpg",1) # BGR olarak okudu
plt.imshow(img) # RGB olarak gösterir
plt.show()
```

```python
# bo sebepten dönüştiriyoz
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img) # RGB olarak gösterir
plt.show()
```

```python
# şimdi de gri formata dönüştürüyoruz
img=cv2.imread("11.2 smile.jpg.jpg",1) # BGR olarak okudu
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray') # cmap : color map # resmin orijinali RGB olmadığından bir işlem daha uygulayacağız
plt.show()
```

```python
img=cv2.imread("11.2 smile.jpg.jpg",1) # BGR olarak okudu
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray',interpolation='BICUBIC') # cmap : color map # resmin orijinali RGB olmadığından bir işlem daha uygulayacağız
plt.show()
```
