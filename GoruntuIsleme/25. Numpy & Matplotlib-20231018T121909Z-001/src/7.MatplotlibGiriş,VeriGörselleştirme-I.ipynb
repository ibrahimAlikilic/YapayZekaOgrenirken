# veri görselleştirme kütüphanesidir

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(5) # 0 dan 5e kadar numpy.array veriyor
y=x
# şimdi grafiğe dökeceğiz
plt.plot(x,y) # grafiğe döktük
plt.show() # bu da göster demek
```

```python
plt.plot(x,y,"*") # ne ile göstermesini istediğimi belirttim
plt.show()
```

```python
plt.plot(x,y,"*-")
plt.show()
```

```python
plt.plot(x,y,"*--") # yıldız yerine o da olur
plt.show()
```

```python
plt.plot(x,y,"*--")
plt.plot(x,-y,)
plt.plot(-x,y,"*-")
plt.title("İlk grafik")
plt.show() # hepsini aynı figure de belrttik
```

```python
N=10
x=np.linspace(0,10,N) # 0dan 10a kadar eşit aralıkta N tane  değerler ver demek
# son tereim eksi ilk terim / artış miktarı +1 deki artış miktarını bilmiyorsun diğer veriler varken ileride kuallanabilirim
print(x)
```

```python
x=np.linspace(0,10,11)
y=x
# eğer grafikte eksenleri görmek istemezsem
plt.plot(x,y,)
plt.axis("off")
plt.show()
# resimleri incelemede bu yöntemi kullancakmışız
```

```python
# eğer grafiği hem x hem y olarak yapmayıp birini tanımlarsam bana index ve değer olarak verir
x=[1,5,7,19,9,17,20]
plt.plot(x)
plt.show()
```

```python
x=[1,5,7,19,9,17,20]
plt.plot(x,[y**2 for y in x]) # y x in içinde dolaşsın ve her bir elemanı liste olarak alsın ben de karesi olarak kullanayım demek
# x**2 yi kabul etmez hata verir
plt.show()
```
