# Histogram 
* Resmin değer yorgunluklarıyla ilgili bilgiler veri
* Aydınlık , karanlık noktaları ve constract değerleri hakkında bilgi verir

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=np.zeros((500,500),np.uint8)
```

```python
# plt.hist() ile histogramı çiziyoruz
# img.ravel() ile 500 , 500 olan şeyi tek bir satır haline dönüştürdük
plt.hist(img.ravel(),256,[0,256]) # kaç değer olduğu , değer aralığı   0,256 0 dan 255e kadar olan renk değerlerim
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.rectangle(img,(0,61),(161,122),(255,255,255),-1) # -1 içi dolu olsun diye
cv2.rectangle(img,(100,20),(200,85),(200,180,161),-1)
cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.hist(img.ravel(),256,[0,256]) # kaç değer olduğu , değer aralığı   0,256 0 dan 255e kadar olan renk değerlerim
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.imread("14.1 smile.jpg.jpg")
plt.hist(img.ravel(),256,[0,256]) # kaç değer olduğu , değer aralığı   0,256 0 dan 255e kadar olan renk değerlerim
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# bgr değerleri için histogram
b,g,r=cv2.split(img)
plt.hist(b.ravel(),256,[0,256]) # kaç değer olduğu , değer aralığı   0,256 0 dan 255e kadar olan renk değerlerim
plt.hist(g.ravel(),256,[0,256])
plt.hist(r.ravel(),256,[0,256])
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
# histogramda en altta kalan ilk yazdırdığım sırasıyla düşün
```
