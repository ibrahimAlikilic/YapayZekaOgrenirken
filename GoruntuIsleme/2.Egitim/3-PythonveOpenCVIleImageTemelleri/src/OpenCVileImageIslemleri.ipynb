```python
# fotoğrafı yeniden boyutlandırma
new_resize=cv2.resize(fix_img,(1800,500))
plt.imshow(new_resize)
```

```python
# şimdi oran ile boyut değiştiricez
# var olanın yarı boyutunda yapalım
w_ratio=0.5
h_ratio=0.5
new_img=cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
print("Eski boyut :",fix_img.shape)
print("Yeni boyut :",new_img.shape)
plt.imshow(new_img)
```

# new_img = cv2.resize(fix_img, (0, 0), fx=w_ratio, fy=h_ratio)
* fix_img: Yeniden boyutlandırmak istediğiniz giriş resmi (NumPy dizisi olarak temsil edilir).* 
* (0, 0): Çıkış boyutunu belirten bir tuple. Bu durumda, çıkış boyutu doğrudan ölçekleme faktörleri ile belirlenir, bu nedenle bu değer (0, 0) olarak bırakılır.* 
* fx=w_ratio: Genişlik için ölçekleme faktörü. Bu faktör, giriş resminin genişliğini w_ratio'ya çarparak yeni genişliği belirler.* 
* fy=h_ratio: Yükseklik için ölçekleme faktörü. Bu faktör, giriş resminin yüksekliğini h_ratio'ya çarparak yeni yüksekliği belirler.

Sonuç olarak, bu fonksiyon, fix_img adlı giriş resmini, belirlenen ölçekleme faktörleri (w_ratio ve h_ratio) kullanılarak yeniden boyutlandırır ve new_img adlı yeni bir resim oluşturur.

```python
# fotoğrafımızı ters çevirme 
new_img=cv2.flip(fix_img,0) # x ekseni için 0
plt.imshow(new_img)
```

```python
new_img=cv2.flip(fix_img,1) # y ekseni için 1
plt.imshow(new_img)
```

```python
# hem x hem y ekseninde çevirmek için
new_img=cv2.flip(fix_img,-1)
plt.imshow(new_img)
```
