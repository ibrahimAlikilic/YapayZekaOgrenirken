```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
bos_foto=np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(bos_foto)
```

```python
cv2.rectangle(bos_foto,(300,100),(400,300),(255,0,0),7)
plt.imshow(bos_foto)
```

```python
cv2.rectangle(bos_foto,(200,200),(300,300),(0,255,0),10) # 1 tane de kare çizelim
plt.imshow(bos_foto)
```

```python
cv2.circle(bos_foto,(100,100),50,(0,0,255),5) # çembe çiziyoruz
plt.imshow(bos_foto)
```

```python
# şimdi de daire çiziyoruz , Bunun için kalınlık değerini - yapıyoruz
cv2.circle(bos_foto,(100,100),50,(0,0,255),-5)
plt.imshow(bos_foto)
```

```python
cv2.line(bos_foto,(0,0),(512,512),(255,255,255),6)
plt.imshow(bos_foto)
```
