```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
bos_foto=np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(bos_foto)
```

```python
cv2.putText(bos_foto,text='SA',org=(10,500),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=2,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA) # org yazının başlangıç konumu
plt.imshow(bos_foto)
```

cv2.putText() fonksiyonu, OpenCV kütüphanesinde bir görüntü üzerine metin yazmak için kullanılır. fontScale parametresi ve lineType parametresi dahil olmak üzere çeşitli parametreler alır.

    fontScale: Bu parametre, fontun ölçeklendirme faktörünü belirtir. Yani, yazı tipi boyutunun ne kadar büyük veya küçük olacağını belirler. Örneğin, 1.0 varsayılan boyuttur, 2.0 yazıyı iki kat daha büyük yapar, 0.5 ise yarı boyutta bir yazı üretir. Bu, metni görüntünün çözünürlüğüne ve istenilen görünürlüğe uygun şekilde ayarlamanıza olanak tanır.

    lineType: Bu parametre, çizilen metnin kenar çizgilerinin tipini belirtir. Yani, metnin kontur çizgilerinin nasıl görüneceğini kontrol eder. Örneğin, cv2.LINE_8, cv2.LINE_AA gibi değerler alabilir. cv2.LINE_AA (anti-aliased line), kenarları daha yumuşak ve daha az dişli görünen metinler için tercih edilir ve genellikle daha kaliteli bir görünüm sağlar. lineType parametresi genellikle metnin okunabilirliğini ve estetik görünümünü iyileştirmek için kullanılır.

### Noktalar belirleyip çokgen çizicez

```python
noktalar=np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
noktalar.shape
```

```python
cv2.polylines(bos_foto,[noktalar],isClosed=True,color=(255,0,0),thickness=5) # isClosed ilk ve son koordinatlardaki noktaları da birleştireyim mi diye soruyor
plt.imshow(bos_foto)
```

```python
# isClosed=False görelim
noktalar=np.array([[100,400],[200,300],[400,400],[200,500]],dtype=np.int32)
cv2.polylines(bos_foto,[noktalar],isClosed=False,color=(0,0,255),thickness=5)
plt.imshow(bos_foto)
```
