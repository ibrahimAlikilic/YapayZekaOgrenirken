Fotoğrafımızın yoğunluğuna ve rengine etki edeceğiz

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# fotoğrafın size sini değiştirmek için fonksiyon yaızyoruz
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,'gray')
```

```python
img=cv2.imread("input/sudokuu.jpg")
display_img(img)
```

## Yata çizgilerin belirgin olması için xGradient uygulayacağız

```python
# depth(derinlik) = duyarlılık olarak düşünelim
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5) # x sobel uygulamak istediğimizden x=1 , y=0 yaptık
display_img(sobelx)
```

## Dikey çizgilerin belirgin olması için yGradient uygulayacağız

```python
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5) # y sobel uygulamak istediğimizden x=1 , y=0 yaptık
display_img(sobely)
```

## Hem yatay hem de dikeyde netlik için Laplacian

```python
laplacian=cv2.Laplacian(img,cv2.CV_64F)
display_img(laplacian)
```

## Hem x hem de y ekseninde uygulamış olduklarımı karıştıracağım bu sayede iki eksen tek fotoğrafta olacak

```python
blended=cv2.addWeighted(sobelx,0.5,sobely,0.5,0)
display_img(blended)
```

```python
# şimdi de threshold uygulayıp biraz daha netleştirelim
ret,frame=cv2.threshold(img,45,255,cv2.THRESH_BINARY)
display_img(frame)
```

```python
# sonucu merak ettim ve eğitimin dışında kendim ekledim
laplacian=cv2.Laplacian(frame,cv2.CV_64F)
display_img(laplacian)
```
