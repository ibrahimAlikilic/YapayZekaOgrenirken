```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("input/kopek.png")
plt.imshow(img)
```

```python
# BGR formatından RGB formatına çeviriyoruz
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# RGB formatından HSV formatına çeviriyoruz
img=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(img)
```

```python
# RGB formatından HSV formatına çeviriyoruz
img=cv2.cvtColor(img,cv2.COLOR_RGB2HLS)
plt.imshow(img)
```
