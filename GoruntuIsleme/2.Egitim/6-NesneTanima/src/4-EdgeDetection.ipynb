```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('input/dog_face.jpg')
plt.imshow(img)
```

```python
# imdi hibir filtre uygulamadan direk canny algoritmasn kullanacaz
edges=cv2.Canny(img,127,127) # ilk nce u ekilde dndk pikseller 0-255 olduundan ortalama deeri threshold olarak kullanalm dedik
plt.imshow(edges) # grdk ki istediimiz verimi alamadk
```

```python
# Daha doru sonu iin ilk nce parazitlerden kurtulalm
blur_img=cv2.blur(img,(5,5))
plt.imshow(blur_img)
```

```python
edges=cv2.Canny(blur_img,90,100) # bu seferki threshold deerlerini byle verelim dedik
plt.imshow(edges)
```

### Yukarda durmadan deer atayp denemektense Trackbar ekleyerek yapalm

```python
import cv2

img = cv2.imread("input/dog_face.jpg")
height,width,_=img.shape # Ubuntuda pencereler kk alyor ve pencerelerin resime orantl olmasn istediimden deerleri aldm
if img is None:
    print("Resim dosyas bulunamad. Ltfen dosya yolunu kontrol edin.")
    exit()

blur_img = cv2.blur(img, (5, 5))

def nothing(x):
    pass

# Gerekli pnecereleir oluturuyprum
cv2.namedWindow("Trackbar")
cv2.namedWindow("edges Uygulanm", cv2.WINDOW_NORMAL)

cv2.resizeWindow("Trackbar", 500, 500)
cv2.createTrackbar("Trackbar 1", "Trackbar", 0, 255, nothing)
cv2.createTrackbar("Trackbar 2", "Trackbar", 0, 255, nothing)



while True:
    cv2.resizeWindow("edges Uygulanm", width*3, height*3)

    trackeba1 = cv2.getTrackbarPos("Trackbar 1", "Trackbar")
    trackeba2 = cv2.getTrackbarPos("Trackbar 2", "Trackbar")


    # Canny kenar alglama ilemi
    edges = cv2.Canny(blur_img, trackeba1, trackeba2)

    cv2.imshow("Orijinal", img)
    cv2.imshow("edges Uygulanm", cv2.resize(edges, (width*3, height*3)))

    key = cv2.waitKey(1)  
    if key == ord("q") or key == 27:  
        break

cv2.destroyAllWindows()
```
