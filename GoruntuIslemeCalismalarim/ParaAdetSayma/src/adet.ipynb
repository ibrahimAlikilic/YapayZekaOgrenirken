```python
# yarıçap için değer aralığı belirterek paraları buldum ve saydırdım .
import cv2
import numpy as np

img1=cv2.imread("include/1.jpeg")
gray1=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img1_blur=cv2.medianBlur(gray1,5) # gürültüleri yok etmek için medyan blur kullanıyoruz

# şimdi çemberleri arayacağız
# resim , algılama yöntemi ( sabit bir şey tek tanımlanmış olan bu var) , çözünürlük oranı = 1 dedik , tespit edilen çemberler arası minimum mesafe genel olarak uygulanan ( img1shape[0]/16), metoda özel girilmesi gereken 2 değer ,tahmini max ve min yariçaplar
circles=cv2.HoughCircles(img1_blur,cv2.HOUGH_GRADIENT,1,img1.shape[0]/16,param1=200,param2=10,minRadius=30,maxRadius=37) 

circle_count=0 # kaç tane para bulduğumuzu saymak için

if circles is not None : 
    circles=np.uint16(np.around(circles)) 
    for i in circles[0,:]: 
        cv2.circle(img1,(i[0],i[1]),i[2],(0,0,255),3)
        circle_count+=1
cv2.putText(img1, f"Cember Sayisi : {circle_count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

# 2. resim için kopyala yapıştır
img2=cv2.imread("include/2.jpeg")
gray2=cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
img2_blur=cv2.medianBlur(gray2,5) 

circles=cv2.HoughCircles(img2_blur,cv2.HOUGH_GRADIENT,1,img2.shape[0]/16,param1=200,param2=10,minRadius=5,maxRadius=21) # /Radius değerlerini değiştirdik

circle_count=0 

if circles is not None : 
    circles=np.uint16(np.around(circles)) 
    for i in circles[0,:]: 
        cv2.circle(img2,(i[0],i[1]),i[2],(0,0,255),3)
        circle_count+=1
cv2.putText(img2, f"Cember Sayisi : {circle_count}", (1, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
cv2.imshow("img1",img1)
cv2.imshow("img2",img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# yarıçap için değer aralığı belirterek paraları buldum ve saydırdım .
import cv2
import numpy as np

img1=cv2.imread("include/1.jpeg")
gray1=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img1_blur=cv2.medianBlur(gray1,5) # gürültüleri yok etmek için medyan blur kullanıyoruz

img2=cv2.imread("include/2.jpeg")
gray2=cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
img2_blur=cv2.medianBlur(gray2,5) 

def cizim(img,img_blur,x,y,yazi_boyuti):
    circles=cv2.HoughCircles(img_blur,cv2.HOUGH_GRADIENT,1,img.shape[0]/16,param1=200,param2=10,minRadius=x,maxRadius=y) 
    circle_count=0 # kaç tane para bulduğumuzu saymak için
    if circles is not None : 
        circles=np.uint16(np.around(circles)) 
        for i in circles[0,:]: 
            cv2.circle(img,(i[0],i[1]),i[2],(0,0,255),3)
            circle_count+=1
    cv2.putText(img, f"Cember Sayisi : {circle_count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, yazi_boyuti, (255, 0, 0), 2)

cizim(img1,img1_blur,30,37,1)
cizim(img2,img2_blur,5,21,0.5)
cv2.imshow("img1",img1)
cv2.imshow("img2",img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Yazdığım kod
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline


def nothing(x):
    pass
# Gerekli pnecereleir oluşturuyorum
cv2.namedWindow("Trackbar")
cv2.namedWindow("Orijinal",cv2.WINDOW_NORMAL)
cv2.namedWindow("thresh",cv2.WINDOW_NORMAL)
cv2.namedWindow("opening",cv2.WINDOW_NORMAL)
cv2.namedWindow("sure_bg",cv2.WINDOW_NORMAL)
cv2.namedWindow("sure_fg",cv2.WINDOW_NORMAL)
cv2.namedWindow("dist_transform",cv2.WINDOW_NORMAL)
cv2.namedWindow("unknown",cv2.WINDOW_NORMAL)
cv2.namedWindow("markers",cv2.WINDOW_NORMAL)
# cv2.namedWindow("markersGri")



cv2.createTrackbar("threshold Min Degeri", "Trackbar", 0, 255, nothing)
cv2.createTrackbar("threshold Max Degeri", "Trackbar", 0, 255, nothing)
cv2.createTrackbar("Kernel Degeri", "Trackbar", 1, 21, nothing)
cv2.createTrackbar("ksize Degeri", "Trackbar", 0, 255, nothing)
cv2.createTrackbar("opening_iterations Degeri", "Trackbar", 0, 33, nothing)
cv2.createTrackbar("sure_bg_iterations Degeri", "Trackbar", 0, 33, nothing)
#############################

# Başlangıç değerini set kısmı
cv2.setTrackbarPos("threshold Min Degeri","Trackbar",0)
cv2.setTrackbarPos("threshold Max Degeri","Trackbar",255)
cv2.setTrackbarPos("Kernel Degeri","Trackbar",3)
cv2.setTrackbarPos("ksize Degeri","Trackbar",35)
cv2.setTrackbarPos("opening_iterations Degeri","Trackbar",2)
cv2.setTrackbarPos("sure_bg_iterations Degeri","Trackbar",3)

# kernal ilk değer atama
kernelDegeri = cv2.getTrackbarPos("Kernel Degeri", "Trackbar")
ksizeDegeri = cv2.getTrackbarPos("Kernel Degeri", "Trackbar")



while True:
    img=cv2.imread("include/4.jpeg") # Resmi içeride tamamlıyorum yoksa her seferinde ekleme olduğundan ilk baştaki resim durmadan eklenmiş halindeki oluyor

    # Pencereleri yeniden boyutlandırma
    cv2.resizeWindow("Trackbar", 500, 500)
    cv2.resizeWindow("Orijinal", 500, 500)
    cv2.resizeWindow("thresh", 500, 500)
    cv2.resizeWindow("opening", 500, 500)
    cv2.resizeWindow("sure_bg", 500, 500)
    cv2.resizeWindow("sure_fg", 500, 500)
    cv2.resizeWindow("dist_transform", 500, 500)
    cv2.resizeWindow("unknown", 500, 500)
    cv2.resizeWindow("markers", 500, 500)
    # cv2.resizeWindow("markersGri", 700, 700)


    # Atamalar
    thresholdMin = cv2.getTrackbarPos("threshold Min Degeri", "Trackbar")
    thresholdMax = cv2.getTrackbarPos("threshold Max Degeri", "Trackbar")
    kernelDegeriTutucu = cv2.getTrackbarPos("Kernel Degeri", "Trackbar")
    # tek olmaları gerektiğinde if kulanıyoruz
    if kernelDegeriTutucu%2!=0:
        kernelDegeri=kernelDegeriTutucu
    ksizeDegeriTutucu = cv2.getTrackbarPos("ksize Degeri", "Trackbar")
    if ksizeDegeriTutucu%2!=0:
        ksizeDegeri=ksizeDegeriTutucu
    opening_iterationsDegeri = cv2.getTrackbarPos("opening_iterations Degeri", "Trackbar")
    sure_bg_iterationsDegeri = cv2.getTrackbarPos("sure_bg_iterations Degeri", "Trackbar")

    # blur
    img=cv2.medianBlur(img,ksize=ksizeDegeri) # ! ksize 35 yerine 25 vermiştim o zmn sol üstteki ve sağ alttakini görmemişti ileriki projelerde bu değer üzerinde de oynarız
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    # threshold
    ret,thresh=cv2.threshold(gray,thresholdMin,thresholdMax,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
    
    # biraz daha gürültü azaltacağım
    kernel=np.ones((kernelDegeri,kernelDegeri),np.uint8)
    opening=cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel,iterations=opening_iterationsDegeri) # morphological transform adı altında geçiyor diğerlerini hatırlamak için bu şekilde araştırabilrsin
   

    # dilate uygulayacağız
    sure_bg=cv2.dilate(opening,kernel,iterations=sure_bg_iterationsDegeri)

    dist_transform=cv2.distanceTransform(opening,cv2.DIST_L2,5)
    ret,sure_fg=cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)

    # fotoğraflara birbirini ekleyeceğim 
    sure_fg=np.uint8(sure_fg)
    unknown=cv2.subtract(sure_bg,sure_fg)

    # yaptığımız aşamalar sonucunda 2li görüntüler elde ettik ve şimdi bunların sayısını öğreneceğiz
    ret,markers=cv2.connectedComponents(sure_fg)
    markers=markers+1 # markers sıfırdan başlıyorlarmış bu yüzden 1 ekledik   
    markers[unknown==255]=0
    # bu bizim katmanları birbirinden ayırma işlemimizi kolaylaştırmış oldu

    markers=cv2.watershed(img,markers)


    # şimdi çizelim 
    contours,hierarchy=cv2.findContours(markers.copy(),cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
    for i in range(len(contours)):
        if hierarchy[0][i][3]==-1:
            cv2.drawContours(img,contours,i,(255,0,0),10)

    cv2.imshow("Orijinal",img)
    # display(thresh,cmap='gray')
    # thresh=cv2.cvtColor(thresh,cv2.COLOR_BGR2GRAY)
    cv2.imshow("thresh",thresh) # zaten siyah beyaz görünümde
    # display(opening,cmap="gray")
    # opening=cv2.cvtColor(opening,cv2.COLOR_BGR2GRAY)
    cv2.imshow("opening",opening)# zaten siyah beyaz görünümde
    # display(sure_bg,cmap="gray")
    # sure_bg=cv2.cvtColor(sure_bg,cv2.COLOR_BGR2GRAY)
    cv2.imshow("sure_bg",sure_bg)# zaten siyah beyaz görünümde
    # display(sure_fg,cmap="gray")
    # sure_fg=cv2.cvtColor(sure_fg,cv2.COLOR_BGR2GRAY)
    cv2.imshow("sure_bg",sure_fg)
    # display(dist_transform,cmap="gray")
    # dist_transform=cv2.cvtColor(dist_transform,cv2.COLOR_BGR2GRAY)
    cv2.imshow("dist_transform",dist_transform)# zaten siyah beyaz görünümde
    # display(unknown,cmap="gray")
    # unknown=cv2.cvtColor(unknown,cv2.COLOR_BGR2GRAY)
    cv2.imshow("unknown",unknown)# zaten siyah beyaz görünümde


    # display(markers)
    # cv2.imshow("markers",markers)
    # # display(markers,cmap="gray")
    # markers=cv2.cvtColor(markers,cv2.COLOR_BGR2GRAY)
    # cv2.imshow("markersGri",markers)

    # `markers` değişkenini uygun formata dönüştürün
    markers = np.uint8(markers)
    
    # `markers` değişkenini renk haritasına dönüştürün
    markers_colored = cv2.applyColorMap(markers, cv2.COLORMAP_JET)
        
    # Renkli markerları gösterin
    cv2.imshow("markers", markers_colored)


    

    key = cv2.waitKey(1)
    if key == ord("q") or key == 27:
        break

cv2.destroyAllWindows()
```
